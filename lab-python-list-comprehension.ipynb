````python
#Function 1:

def initialize_inventory(products):
    inventory = {}
    for product in products:
        number_of_items = int(input(f"How many {product} are there in the inventory?"))
        inventory[product] = number_of_items
    return inventory

products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = initialize_inventory(products) # creates the inventory from the "products" list with quantities from user input

print(inventory)

comprehended_inventory = {
    product: int(input(f"How many {product} are there in the inventory?"))
    for product in products
}
````

````python
#Function 3:

def update_inventory(customer_orders,inventory):
    updated_inventory = {}                                          #creates a new empty dict 
    for product in inventory.keys():                                #loops run through the products (keys) in the current inventory 
        if product in customer_orders:
            updated_inventory[product] = int(inventory[product] - 1) #only subtracts 1 from customer orders, and updates the values into the new dict
        else:
            updated_inventory[product] = inventory[product]          #if the products were not purchased, the values remain the same
    return updated_inventory

updated_inventory = update_inventory(customer_orders, inventory)
print(updated_inventory)

comprehended_inventory = {
    product: inventory[product] - 1 if product in customer_orders else inventory[product]
    for product in inventory
}
````

````python
#Function 2:

def get_customer_orders():
    number_of_orders = int(input("How many products does the customer want to order? "))

    customer_orders = {
        order
        for product in range(number_of_orders)
        for order in [input(f"Enter product {product+1}: ").strip().lower()]
        if order in products
    }

    return customer_orders

products = ["t-shirt", "mug", "hat", "book", "keychain"]
customer_orders = get_customer_orders()
print(customer_orders)
````

````python
#New function: calculate total price

def calculate_total_price(customer_orders):
    prices = [
        float(input(f"Enter the price for {product}: "))
        for product in customer_orders
    ]
    total_price = sum(prices)
    return total_price


total = calculate_total_price(customer_orders)
print(f"Total price of the order: ${total}")
````

````python
#regular

def update_inventory(customer_orders, inventory):
    updated_inventory = inventory.copy()
    
    for product, quantity in customer_orders.items():
        if product in updated_inventory:
            updated_inventory[product] -= quantity
            if updated_inventory[product] <= 0:
                updated_inventory.pop(product)
    
    return updated_inventory

#comprehended

def update_inventory(customer_orders, inventory):
    updated_inventory = {
        product: inventory[product] - customer_orders.get(product, 0)
        for product in inventory
    }
    
    #filter out products with quantity == 0
    updated_inventory = {
        product: quantity
        for product, quantity in updated_inventory.items()
        if quantity != 0
    }
    
    return updated_inventory
````
